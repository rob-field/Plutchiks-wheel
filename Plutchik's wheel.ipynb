```python
from pyplutchik import plutchik
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Random Data
emotions = {'joy': 0.6,
            'trust': 0.4,
            'fear': 0.1,
            'surprise': 0.7,
            'sadness': 0.1,
            'disgust': 0.5,
            'anger': 0.4,
            'anticipation': 0.6}
```

```python
fig, ax = plt.subplots( nrows = 1, ncols = 1, figsize = (15, 15))

plutchik(emotions,
         ax = ax,
         font = None, 
         fontweight = 'light',
         fontsize = 15, 
         show_coordinates = True, 
         show_ticklabels = False, 
         highlight_emotions = 'all', 
         show_intensity_labels = 'none',
         ticklabels_angle = 0, 
         ticklabels_size = 11, 
         height_width_ratio = 1, 
         title = None, 
         title_size = None, 
         normalize = False )
```

```python
# Read in data and convert to dictionary
# [,8:] contain the dyads so exclude for now
df = pd.read_excel("Plutchik's Wheel Outputs.xlsx", "python_data")
emotion_dict = df.iloc[:,:8].to_dict('list')
emotion_dict
```

```python
fig, ax = plt.subplots( nrows = 1, ncols = 1, figsize = (15, 15))

plutchik(emotion_dict,
         ax = ax)
```

```python
fig, ax = plt.subplots( nrows = 1, ncols = 1, figsize = (15, 15))

plutchik(emotion_dict,
         ax = ax,
         show_intensity_labels = ['anticipation', 'trust', 'anger'],
         show_ticklabels = True, 
         ticklabels_angle = 28)
```

```python
fig, ax = plt.subplots( nrows = 1, ncols = 3, figsize = (26, 8) )

plutchik(emotion_dict, ax = ax[0])
plutchik(emotion_dict, ax = ax[1], highlight_emotions = ['anticipation', 'trust'])
plutchik(emotion_dict, ax = ax[2], highlight_emotions = ['anticipation', 'trust'], show_intensity_labels = ['anticipation', 'trust'])
```
